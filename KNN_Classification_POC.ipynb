**x y split**

```python
x = df.iloc[:,0:7]  # all rows, 0-7th index are x values
x
```

```python
y = df.iloc[:,-1]  # all rows, last column is y
y
```

**TRAIN TEST SPLIT**

After splitting the dataset into x and y, we are going to train the dataset first and then test the dataset. For that we need to import train_test_split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.25, random_state = 22)
```

***MODEL BUILDING***

**KNN - K NEAREST NEIGHBOR**

KNN is a simple, supervised machine learning (ML) algorithm that can be used for classification or regression task.

KNN will compare the new data point with the neighbors and assign it to a class.

K is the number of nearest neighbors to use.

Here, we will see how to implement the KNN algorithm for classification and evaluation of the result.

```python
y.unique # to determine the number of classifiers in target variable y.
```

Since 0 and 1 are the only two unique elements in y variable, this is a binary classifier problem. Hence we must use p = 2.

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, p = 2)
classifier.fit(x_train,y_train)
```

```python
#predicting the test set result
y_pred = classifier.predict(x_test)
y_pred
```

```python
#Creating confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
```

```python
cm  # accuracy = 104+33 = 137
```

```python
from sklearn.metrics import accuracy_score
accuracy_score = accuracy_score(y_test, y_pred)
accuracy_score
```
